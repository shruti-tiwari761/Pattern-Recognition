# Scatter Graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Adjusting y to have the same number of elements as x
x = np.arange(0, 10)  # 10 elements
y = np.arange(11, 21)  # Changed to 11 to 20 to have 10 elements

plt.scatter(x, y, c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.show()
```

```python
plt.plot(x,y,'k')
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'r*--')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Adjusting y to have the same number of elements as x
x = np.arange(0, 10)  # 10 elements
y = np.arange(11, 21)
plt.plot(x,y,'g*',linestyle='dashed',linewidth=1,markersize=20)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

# Creating Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
plt.subplot(3,3,1)
plt.plot(x,y,'r--')
plt.subplot(3,3,2)
plt.plot(x,y,'g*',linestyle='dashed',linewidth=1,markersize=10)
plt.subplot(3,3,3)
plt.plot(x,y,'bo')
plt.subplot(3,3,4)
plt.plot(x,y,'go')
plt.subplot(3,3,5)
plt.plot(x,y,'k--')
plt.subplot(3,3,6)
plt.plot(x,y,'r*--')
plt.subplot(3,3,7)
plt.plot(x,y,'yo')
plt.subplot(3,3,8)
plt.plot(x,y,'bo')
plt.subplot(3,3,9)
plt.plot(x,y,'g--')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib Demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
#compute the x and y coordinates for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.title("cosine wave form")
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.tan(x)
plt.title("tan wave form")
plt.plot(x,y)
```

```python
#Subplot()
#Compute the x and y coordinate for points on sine and cosine
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title("Sine")
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g*-')
plt.title("Cosine")
plt.show()
```

# Bar Graph

```python
x=[2,8,10]
y=[11,16,4]
x2=[3,6,72]
y2=[7,56,23]
plt.bar(x,y, color= 'k')
plt.bar(x2,y2,color='r')
plt.title("Bar Graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

# Histogram

```python
a=np.array([22,86,56,3,123,67,89,23])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
labels=['Python','C++','Ruby','Java']
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightgreen','pink']
explode=(0,0.5,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
