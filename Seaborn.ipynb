# Seaborn Library

# Introduction to relational/ statistical plot
1)scatter plot
2)Line plot
3)Facetgrids

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
import seaborn as sns
df=sns.load_dataset("tips")
df.head()
```

```python
# plot a scatter plot between total bill and tip
```

```python
sns.relplot(x='total_bill',y='tip', data=df)
```

```python
# IF we want scatter plot b/w numerical and categori
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker', data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex', data=df)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex', size='size', data=df)
```

# Using Scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',style='sex', size='size', data=df)
```

# Join plot 
it allow you to study b/w 2 numerical data

```python
sns.jointplot(x='total_bill',y='tip', data=df)
```

```python
sns.jointplot(x='total_bill',y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='reg')
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

# Dist Plot

```python
import seaborn as sns
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
sns.distplot(df['tip'],kde=False,bins=100)
```

```python
sns.distplot(df['tip'],hist=False)
```

```python
sns.distplot(df['tip'],kde=False,hist=False, rug=True)
```
